```python
##IMPORTANT
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
stopwords=set(STOPWORDS)
print(stopwords)
wc=WordCloud(width=800,height=400,max_words=50,background_color='white')
s='B6 is best.Vishal vishal vishal b6 vishal b6 vishal b6 b1 is best b3 is best cs is bestn python FSD DE COA TOC'
wc.generate(s)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
