```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
d={'x':[1,2,3,4,5,6],
    'y1':[10,30,50,70,80,110],
  'y2':[40,70,20,90,30,70]}
df=pd.DataFrame(d)
print(df)
plt.stackplot(df.x,df.y1,df.y2,labels=['y1','y2'])
plt.title('vishal10')
plt.legend()
plt.show()
```

```python
plt.fill_between(df.x,df.y1,df.y2)
```

```python
plt.fill_between(df.x,df.y1,df.y2)
plt.plot(df.x,df.y1,'o:b')
plt.plot(df.x,df.y1,'v-r')
```

```python
##IMPORTANT
plt.fill_between(df.x,df.y1,df.y2,where=df.y1>df.y2,color='red',interpolate=True)
plt.fill_between(df.x,df.y1,df.y2,where=df.y1<df.y2,color='green',interpolate=True)
plt.show()
```

```python
d={'x':[10,30,60,70,80,120],
    'y1':[10,30,50,70,80,110],
  'y2':[40,70,20,90,30,70]}
df=pd.DataFrame(d)
plt.scatter(df.x,df.y1)
plt.scatter(df.x,df.y2)
```

```python
d={'x':[10,30,60,70,80,120],
    'y1':[10,30,50,70,80,110],
  'y2':[40,70,20,90,30,70]}
df=pd.DataFrame(d)
plt.scatter(df.x,df.y1,marker='v',s=30)
plt.scatter(df.x,df.y2,marker='x',s=70)
```
